## Application of Inference Attack Technique

```python
import pandas as pd

# read the previously created datasets
df_is = pd.read_csv("data/is_projects_results.csv")
df_ersu = pd.read_csv("data/ersu_results.csv")
df_alma = pd.read_csv("data/almalaurea_report.csv")
df_outlook = pd.read_csv("data/msoutlook_report.csv")
```

```python
# merge IS dataset with msoutlook dataset on pseudonym field
df_merge1 = pd.merge(df_is, df_outlook, on="Pseudonym")
df_merge1
```

```python
# filter female students
df_merge1 = df_merge1.query("Sex == 'M'")
# filter students that failed the project
df_merge1= df_merge1.query("Passed == 1")
df_merge1
```

```python
# merge the previous dataset with almalaurea dataset on pseudonym field
df_merge2 = pd.merge(df_merge1, df_alma, on="Pseudonym")
df_merge2
```

```python
# filter not off-site students
df_merge2 = df_merge2.loc[df_merge2["Off-site"] == 1]
df_merge2
```

```python
# merge the previous dataset with ersu dataset
df_merge3 = pd.merge(df_merge2, df_ersu, on="Pseudonym")
df_merge3
```

```python
# save pseudonyms that match Bob's profile
df_merge3.to_csv("data/filtered_students.csv", index=False)
```
